# Task 1.1 – The 'Discovery' Architecture (Goorha & Ungar, 2010)

This notebook describes the four-step **Discovery** pipeline introduced by Goorha & Ungar (2010) for identifying **significant emerging trends** in text corpora (e.g., Twitter).

## Step 1: Entity Extraction

**Goal:** Identify meaningful *n*-grams (unigrams and bigrams) from the raw tweet stream that serve as candidate trend entities.

- The system tokenises each tweet and extracts all unigrams and bigrams.
- Stop-words and purely punctuation tokens are discarded.
- The result is a large vocabulary of candidate phrases that will be evaluated in subsequent steps.

> *Why this matters:* Raw tweets are noisy. Limiting candidates to single and two-word phrases keeps the vocabulary tractable while still capturing the colloquial expressions most likely to represent emerging topics.

## Step 2: Proximity Filtering (< 100 characters)

**Goal:** Restrict analysis to *co-occurring* entity pairs that appear **within 100 characters** of each other inside the same tweet.

- For every tweet, pairs of extracted entities are formed.
- A pair is retained only if the character distance between the two entities is **less than 100 characters**.
- Pairs that exceed this threshold are discarded as they are unlikely to be semantically related.

> *Why this matters:* Short-range co-occurrence is a strong proxy for semantic relatedness. The 100-character cutoff is empirically motivated by the median tweet length (~140 characters at the time) and ensures that only genuinely proximate – and therefore plausibly related – entity pairs are scored.

## Step 3: Exploding Trend Detection (50 % increase over 3-week baseline)

**Goal:** Flag entities whose frequency has **increased by at least 50 %** relative to a rolling 3-week historical baseline.

- A 3-week (21-day) moving-average baseline frequency is computed for each entity.
- The entity's current-period frequency is compared with this baseline.
- An entity is labelled an *exploding trend* if:

$$\frac{f_{\text{current}} - f_{\text{baseline}}}{f_{\text{baseline}}} \geq 0.50$$

- Entities below this threshold are pruned from further consideration.

> *Why this matters:* The 50 % threshold separates genuine spikes from ordinary day-to-day variance. The 3-week window is long enough to smooth out weekly seasonality yet short enough to capture fast-moving trends before they peak.

## Step 4: Power-Law Scoring (exponent = 0.95)

**Goal:** Rank surviving trend candidates by a **power-law-weighted** score that rewards genuine spikes while dampening the advantage of already high-frequency terms.

- Each candidate entity is scored using:

$$\text{score}(e) = f(e)^{0.95}$$

  where $f(e)$ is the current (normalised) frequency of entity $e$.

- An exponent of **0.95** (slightly sub-linear) compresses the raw frequency scale, so that extremely common terms do not dominate over genuinely novel, rapidly-growing terms.
- Entities are then ranked in descending order of score; the top-ranked entities are reported as the discovered trends.

> *Why this matters:* A purely linear frequency ranking would favour perennially popular terms. The sub-linear exponent of 0.95 achieves the best empirical signal-to-noise ratio by giving a small but meaningful penalty to high-baseline entities and a slight relative boost to newer, explosively growing ones.
