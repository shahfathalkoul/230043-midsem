# Task 1.1 — Core Architecture of "Discovery of Significant Emerging Trends"

**Paper:** Goorha, S. & Ungar, L. (2010). *Discovery of Significant Emerging Trends.* ACM SIGKDD.

---

## Step-by-Step Method Description

The Discovery architecture identifies significant emerging trends from large-scale text corpora (e.g., blog posts, news articles) through a four-stage pipeline:

### Step 1: Entity Extraction (Identifying Brands)

The first stage scans the raw text corpus to **extract named entities** — specifically brand names, product names, and other proper nouns. This stage converts unstructured text into a structured set of entity mentions with their positions in the document. The goal is to build a comprehensive inventory of *what* is being discussed across the corpus.

### Step 2: Proximity Filtering (< 100 Character Window)

Once entities are extracted, the system identifies **co-occurring phrase pairs** that appear within a **100-character window** of each other. The proximity constraint acts as a relevance heuristic: if two terms appear close together in text, they are likely semantically related. This step dramatically reduces the combinatorial space of possible phrase pairs, filtering out spurious long-distance co-occurrences that are unlikely to represent meaningful associations.

### Step 3: Exploding Trend Detection (50% Increase Over 3-Week Baseline)

For each surviving phrase pair, the system tracks its **frequency over time** and compares the current count against a **rolling 3-week baseline**. A phrase pair is flagged as an "exploding trend" if its current frequency shows a **≥ 50% increase** relative to the baseline. This temporal filter ensures that only genuinely *emerging* associations are surfaced, filtering out stable, long-standing co-occurrences.

### Step 4: Power-Law Scoring

Finally, each exploding trend is ranked using the **power-law scoring formula**:

$$\text{Score} = \frac{\text{count}}{\text{total}^{0.95}}$$

where:
- `count` = the number of times this specific phrase pair appears
- `total` = the total number of all phrase pair occurrences
- `0.95` = the scaling exponent that balances volume and specificity

The sub-linear exponent (0.95 < 1.0) **down-weights high-frequency generic phrases** while still giving credit to volume. This prevents ubiquitous terms like "the" or "and" from dominating the rankings, while ensuring that genuinely popular emerging trends are ranked higher than rare noise.

---

**Summary:** The Discovery architecture is a four-stage pipeline that extracts entities, filters by proximity, detects temporal explosions, and ranks by a power-law score — transforming raw text into a ranked list of significant emerging trends.
