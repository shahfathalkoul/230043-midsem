# Task 1.2 – Key Assumptions of the Discovery Architecture

The **Discovery** pipeline (Goorha & Ungar, 2010) rests on three foundational assumptions that shape its design choices. Each assumption is examined below.

## Assumption 1: Proximity Implies Relevance

The architecture assumes that two entities appearing **within 100 characters** of each other in a tweet are semantically related.

- Tweets are short, informal utterances; authors rarely mention unrelated concepts in close proximity.
- Character-level proximity is therefore treated as a reliable, low-cost proxy for semantic co-relevance.
- This assumption allows the system to build entity-pair signals without the expense of full syntactic or semantic parsing.

**Implication:** If two trending terms co-occur frequently and proximately, the system interprets their joint signal as evidence of a single, coherent emerging topic rather than two independent ones.

## Assumption 2: Trends Must Persist Beyond 'One-Day Wonders'

The architecture assumes that a **genuine trend** is one whose elevated frequency is sustained, not merely a single-day spike driven by a transient event.

- The 3-week rolling baseline is designed to distinguish persistent growth from isolated bursts.
- A term that spikes for one day and then returns to its baseline fails the 50 % threshold when measured over the full window.
- Only terms that maintain elevated frequency across multiple observation periods are classified as real emerging trends.

**Implication:** The system deliberately sacrifices sensitivity to flash-in-the-pan events in order to surface durable, structurally significant trends that are more useful to analysts and decision-makers.

## Assumption 3: The 0.95 Exponent Creates the Best Signal-to-Noise Ratio

The architecture assumes that a **sub-linear power-law exponent of 0.95** is the empirically optimal choice for balancing frequency signal against the noise introduced by high-baseline popularity.

- A linear exponent (1.0) would simply rank by raw frequency, over-rewarding already-popular terms whose growth rate is unremarkable.
- A highly sub-linear exponent (e.g., 0.5) would over-level the playing field, inflating the scores of very rare terms that may be statistically insignificant.
- At **0.95**, the compression is gentle enough to preserve the signal from genuinely frequent-but-growing terms while slightly penalising entrenched high-frequency vocabulary.

**Implication:** The choice of 0.95 is not theoretically derived; it is an empirically tuned hyperparameter. This means the assumption holds only for corpora with distributional properties similar to those used in the original study (Twitter, 2009–2010). Transfer to very different corpora may require re-tuning the exponent.
