# Task 1.2 — Key Assumptions of the Discovery Method

**Paper:** Goorha, S. & Ungar, L. (2010). *Discovery of Significant Emerging Trends.* ACM SIGKDD.

---

The Discovery architecture rests on three critical assumptions that underpin its design choices:

## Assumption 1: Proximity Implies Relevance

The method assumes that if two phrases appear within a **100-character window**, they are **semantically related**. This is a strong heuristic: physical proximity in text is used as a proxy for topical association. While this works well in structured news articles and blog posts, it may break down in contexts with complex sentence structures, sarcasm, or interleaved topics. The 100-character threshold was chosen empirically by the authors as a practical trade-off between capturing genuine associations and excluding noise from distant co-occurrences.

## Assumption 2: Trends Must Persist (No "One-Day Wonders")

The exploding trend detection requires a **≥ 50% increase over a 3-week rolling baseline**. This implicitly assumes that **genuine trends persist over multiple days**, not just spike for a single day. One-day bursts — such as a celebrity gaffe that trends for 24 hours and then vanishes — are deliberately filtered out. The 3-week baseline window smooths out daily volatility and ensures that only sustained upward movements are flagged. This design choice reflects the paper's focus on *significant* trends rather than transient viral events.

## Assumption 3: 0.95 Is the Optimal Scaling Exponent

The power-law scoring formula uses an exponent of **0.95**, positioned between two extremes:
- **α = 0.0 (Pure Volume):** Score = count — high-frequency generic words dominate
- **α = 1.0 (Pure Specificity):** Score = count/total — rare junk phrases dominate

The authors assume that **0.95 is the optimal trade-off** between these extremes. This near-linear but slightly sub-linear exponent heavily penalises the denominator (total occurrences), favouring phrases that are common enough to be meaningful but not so common as to be noise. The specific value 0.95 was determined empirically by the authors through experimentation on their news corpus, and may not generalise perfectly to all domains or corpus sizes.

---

**Summary:** The three assumptions — proximity as relevance, trend persistence, and the 0.95 exponent — are the foundational pillars of the Discovery method. Each represents a pragmatic trade-off that works well in the paper's target domain but may require recalibration for other contexts.
